```python
import os
import sys
import json
import math
import random
import datetime
import collections

# Very long and repetitive way to compute squares
result_list = []
for i in range(0, 10, 1):
    val = i * i
    result_list.append(val)

# Duplicated logic instead of a function
total_a = 0
for num in result_list:
    total_a = total_a + num

total_b = 0
for num in result_list:
    total_b = total_b + num

print(total_a, total_b)
```

```python
# Inefficient string concatenation in a loop
words = ['hello', 'world', 'this', 'is', 'python']
sentence = ''
for w in words:
    sentence = sentence + w + ' '
sentence = sentence.strip()

# Verbose conditional that can be simplified
x = 42
if x > 0:
    is_positive = True
else:
    is_positive = False

# Unnecessary temporary variables
temp1 = len(words)
temp2 = temp1 * 2
temp3 = temp2 + 1
final = temp3

print(sentence, is_positive, final)
```

```python
# Nested loops that could use zip or enumerate
names = ['Alice', 'Bob', 'Charlie']
scores = [95, 87, 92]
i = 0
pairs = []
while i < len(names):
    pair = (names[i], scores[i])
    pairs.append(pair)
    i = i + 1

# Manual max without built-in
max_score = scores[0]
for s in scores:
    if s > max_score:
        max_score = s

print(pairs, max_score)
```
